```python
import hashlib
import os

# Example malware hash database
# Replace these with real SHA-256 hashes if needed.
malware_hashes = {
    "e3b0c44298fc1c149afbf4c8996fb92427ae41e4649b934ca495991b7852b855",
    "5d41402abc4b2a76b9719d911017c592"
}

def calculate_sha256(file_path):
    sha256 = hashlib.sha256()

    with open(file_path, "rb") as file:
        while chunk := file.read(4096):
            sha256.update(chunk)

    return sha256.hexdigest()

file_path = input("Enter file path: ")

if os.path.exists(file_path):
    file_hash = calculate_sha256(file_path)

    print("\nSHA-256 Hash:")
    print(file_hash)

    if file_hash in malware_hashes:
        print("\n⚠ Malware Detected")
    else:
        print("\n✓ File appears safe")
else:
    print("File not found.")
```

```python
from urllib.parse import urlparse
import re

def detect_phishing(url):

    score = 0

    if len(url) > 75:
        score += 1

    if "@" in url:
        score += 1

    if "-" in urlparse(url).netloc:
        score += 1

    if not url.startswith("https://"):
        score += 1

    ip_pattern = r'http[s]?://(\d+\.){3}\d+'

    if re.match(ip_pattern, url):
        score += 2



    if score >= 3:
        print("⚠ Possible Phishing Website")
    else:
        print("✓ Looks Legitimate")

url = input("Enter URL: ")
detect_phishing(url)
```

```python
from collections import Counter

LOG_FILE = "server_log.txt"

THRESHOLD = 5

with open(LOG_FILE, "r") as file:
    ips = file.read().splitlines()

counts = Counter(ips)

print("\nTraffic Summary\n")

for ip, count in counts.items():

    print(f"{ip} -> {count} requests")

    if count >= THRESHOLD:
        print("⚠ Possible DoS Attack")
```

```python
from scapy.all import sniff

def process_packet(packet):

    if packet.haslayer("IP"):

        print("--------------------------")

        print("Source:", packet["IP"].src)

        print("Destination:", packet["IP"].dst)

        print("Protocol:", packet["IP"].proto)

sniff(prn=process_packet, count=20)
```

```python
import socket

target = input("Enter IP Address: ")

ports = [20,21,22,23,25,53,80,110,143,443,3306,3389]

print("\nScanning...\n")

for port in ports:

    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

    sock.settimeout(1)

    result = sock.connect_ex((target, port))

    if result == 0:
        print(f"Port {port} : OPEN")

    sock.close()

print("\nScan Complete")
```
